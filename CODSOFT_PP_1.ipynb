```python
def show_menu():
    print("Todo List Application")
    print("1. View Todo List")
    print("2. Add Todo")
    print("3. Mark Todo as Done")
    print("4. Delete Todo")
    print("5. Exit")


def view_todo_list(todo_list):
    if not todo_list:
        print("Your todo list is empty!")
    else:
        print("Todo List:")
        for idx, item in enumerate(todo_list, start=1):
            print(f"{idx}. {item}")


def add_todo(todo_list):
    todo_item = input("Enter the task you want to add: ")
    todo_list.append(todo_item)
    print("Task added successfully!")


def mark_todo_done(todo_list):
    view_todo_list(todo_list)
    if todo_list:
        idx = int(input("Enter the index of the task you want to mark as done: ")) - 1
        if 0 <= idx < len(todo_list):
            print(f"Marking '{todo_list[idx]}' as done.")
            todo_list.pop(idx)
        else:
            print("Invalid index!")
    else:
        print("No tasks to mark as done!")


def delete_todo(todo_list):
    view_todo_list(todo_list)
    if todo_list:
        idx = int(input("Enter the index of the task you want to delete: ")) - 1
        if 0 <= idx < len(todo_list):
            print(f"Deleting '{todo_list[idx]}'")
            todo_list.pop(idx)
        else:
            print("Invalid index!")
    else:
        print("No tasks to delete!")


def main():
    todo_list = []

    while True:
        show_menu()
        choice = input("Enter your choice: ")

        if choice == "1":
            view_todo_list(todo_list)
        elif choice == "2":
            add_todo(todo_list)
        elif choice == "3":
            mark_todo_done(todo_list)
        elif choice == "4":
            delete_todo(todo_list)
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please try again.")


if __name__ == "__main__":
    main()
```
